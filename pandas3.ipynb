```python
df_sale=pd.read_csv(r'D:\pandas data set\fsdl data2\sales_data_final (1).csv')
```

```python
df_sale.head()
```

```python
def profit_flag(a):
    if a>=0:
        return "positive"
    else:
        return"negative"
```

```python
profit_flag(-45)
```

```python
profit_flag(45)
```

```python
df_sale['flag_profit']=df_sale['profit'].apply(profit_flag)
```

```python
df_sale.head()
```

```python
df_sale['len_cust_name']=df_sale['customer_name'].apply(len)
```

```python
df_sale.head()
```

```python
def quantity_flag(a):
    if a<10 :
        return 'low'
    elif a>10 and a<20 :
        return 'medium'
    else:
        return 'high'
```

```python
df_sale['flag_quantity']=df_sale['quantity'].apply(quantity_flag)
```

```python
df_sale.head()
```

```python
df_sale['square_quantity']=df_sale['quantity'].apply(lambda a: a**2)
```
